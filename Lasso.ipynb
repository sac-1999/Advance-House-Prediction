```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
house = pd.read_csv('house.csv')
test = pd.read_csv('test_data.csv')
```

```python
cormat = house.corr()
cormat[cormat>0.8]

```

```python
x = house.drop('SalePrice',axis=1)
y = house.loc[:,'SalePrice']


from sklearn.linear_model import LassoCV
selection  = LassoCV()
selection.fit(x, y)

selection_score = pd.DataFrame(selection.coef_, index = x.columns)
final_features = list(selection_score[selection_score[0]>0].index)

x = x[final_features]



from sklearn.model_selection import train_test_split
x_train, x_test , y_train, y_test = train_test_split(x,y, test_size= 0.2)
x_train
```

Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
x_train = minmax.fit_transform(x_train)
x_test = minmax.fit_transform(x_test)
```

Feature Selection 

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso()
lasso_reg.fit(x_train,y_train)

lasso_pred = lasso_reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score

r2_score(lasso_pred,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'alpha': [0.001, 0.01, 0.1, 0, 1, 10, 50, 100,1000]}
gridsearch = GridSearchCV(lasso_reg, param_grid, cv=5, verbose=2, scoring='r2').fit(x, y)
```

```python
gridsearch.best_params_
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha = 100)
lasso_reg.fit(x_train,y_train)

lasso_pred = lasso_reg.predict(x_test)
from sklearn.metrics import r2_score
r2_score(y_test, lasso_pred)
```

```python
lasso_pred = pd.DataFrame(lasso_pred,index = test.index, columns = ['SalePrice'])
lasso_pred = pd.concat([test[['Id']],lasso_pred],axis=1)

from sklearn.metrics import r2_score
r2_score(lasso_pred,y_test)
```

```python
lasso_pred.to_csv('lasso.csv', index=False)
```
